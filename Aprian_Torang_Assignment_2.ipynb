#Q3

How to get the rows of a dataframe with row sum > 100?

```python
df2 = pd.DataFrame(np.random.randint(10, 40, 60).reshape(-1, 4))
df2
```

```python
sum_row = df2.sum(axis = 1)
df2[4] = sum_row
df2
```

```python
result = np.where(df2[4] == 100)
row_num = result[0][0] + 1
print(f"the row of value 100 is {row_num}")
```
